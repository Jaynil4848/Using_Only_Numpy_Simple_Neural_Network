```python
def ReLU(x):
    return(np.maximum(0.0,x))

def Softmax(z):
##    e = np.exp(z)
##    return(e/np.sum(e))
  max_z = np.max(z, axis=1, keepdims=True)
  z_exp = np.exp(z - max_z)
  sum_z_exp = np.sum(z_exp, axis=1, keepdims=True)
  return z_exp / sum_z_exp
def deriv_ReLU(z):
    return z>0
```

```python
def forward(w1,w2,b1,b2):
   ## super().init_params()

    ## For the Hidden Layer 1
    z1 = w1.dot(x_train) + b1
    ##RelU Activation Function 
    a1 = ReLU(z1)
    
    ## For the Hidden Layer 2
    z2 = w2.dot(z1) + b2
    a2 = Softmax(z2)
    
    return z1,a1,z2,a2
```

```python
def one_hot(Y):
    max_val = int(Y.max())  # Cast the maximum value of Y to an integer
    one_hot_Y = np.zeros((Y.size, max_val + 1))
    one_hot_Y[np.arange(Y.size), Y.astype(int)] = 1
    one_hot_Y = one_hot_Y.T  # Transpose the result if needed
    return one_hot_Y
```

```python
def backward(z1,a1,z2,a2,y_train,x_train,w1,w2):
    y = one_hot(y_train)
    dz2 = a2 - y ## dz2 here is the error in the output of the second layer  
    dw2 = (1/m)*dz2.dot(a1.T) ## Here we are changing the weights of the final ouput layer  
    db2 = np.sum(z2)*(1/m)
    
    dz1 = w2.T.dot(dz2)*deriv_ReLU(z1) ## changing the params of the 1st layer 
    dw1 = (1/m)*dz1.dot(x_train.T)
    db1 = (1/m)*np.sum(dz2)
    
    return dw1,dw2,db1,db2
    
```

```python
def update_params(dw1,dw2,db1,db2,w1,w2,b1,b2,alpha):
    w1 = w1 - alpha*dw1
    w2 = w2 - alpha*dw2
    b2 = b2 - alpha*db2
    b1 = b1 - alpha*db1
    return w1,w2,b1,b2
```
